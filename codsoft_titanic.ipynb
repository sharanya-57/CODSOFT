```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
titanic_df=pd.read_csv("Titanic-Dataset.csv")
print(titanic_df)
```

```python
# Explore the dataset
print(titanic_df.head())  
```

```python
print(titanic_df.describe())
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
```

```python
titanic_df = titanic_df.drop(columns=['Cabin'])
```

```python
print(titanic_df.isnull().sum())
```

```python
print(titanic_df.columns)  
```

```python
print(titanic_df.dtypes)
```

```python
# Visualize the distribution of numerical features
sns.histplot(titanic_df['Age'], kde=True,color='purple')
plt.title('Distribution of Age')
plt.show()
```

```python
# Visualize survival rate by passenger class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival rate by embarkation port
sns.countplot(x='Embarked', hue='Survived', data=titanic_df)
plt.title('Survival Count by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival rate by gender
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0,1], labels=['Male', 'Female'])
plt.show()
```

```python
# Visualize age distribution by survival
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival rate by number of parents/children
sns.countplot(x='Parch', hue='Survived', data=titanic_df)
plt.title('Survival Count by Number of Parents/Children')
plt.xlabel('Number of Parents/Children')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival rate by number of siblings/spouses
sns.countplot(x='SibSp', hue='Survived', data=titanic_df)
plt.title('Survival Count by Number of Siblings/Spouses')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival rate by embarkation port and passenger class
sns.catplot(x='Embarked', hue='Survived', col='Pclass', data=titanic_df, kind='count')
plt.suptitle('Survival Count by Embarkation Port and Ticket Class')
plt.subplots_adjust(top=0.85)
plt.show()
```

```python
# Visualize survival rate by ticket class and gender
sns.catplot(x='Pclass', hue='Survived', col='Sex', data=titanic_df, kind='count')
plt.suptitle('Survival Count by Ticket Class and Gender')
plt.subplots_adjust(top=0.85)
plt.show()
```

```python
# Explore relationships between features and survival using visualizations 
sns.pairplot(titanic_df, hue='Survived')
plt.show()
```
